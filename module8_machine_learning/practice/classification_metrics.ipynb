```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
cancer = pd.read_csv("/home/darkstar/Documents/pg-dbda/module7_statistics/Daywise Study Material/cases/Cancer/Cancer.csv")
cancer
```

```python
dumm_cancer = pd.get_dummies(cancer, drop_first=True)
dumm_cancer = dumm_cancer.astype(int)
dumm_cancer
```

```python
X = dumm_cancer.drop(['subjid', 'Class_recurrence-events'], axis=1)
y = dumm_cancer['Class_recurrence-events']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
nb = BernoulliNB()
nb.fit(X_train, y_train)


y_pred = nb.predict(X_test)
print(confusion_matrix(y_pred, y_test))

```

## Glass Dataset

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, log_loss
```

```python
glass = pd.read_csv("/home/darkstar/Documents/pg-dbda/module7_statistics/Daywise Study Material/cases/Glass Identification/Glass.csv")
glass
```

```python
X = glass.drop(['Type'], axis=1)
y = glass['Type']

lb1 = LabelEncoder()
y = lb1.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=23)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)

y_pred = nb.predict(X_test)
y_pred_prob = nb.predict_proba(X_test)
print(classification_report(y_pred, y_test))
print("log loss: ",log_loss(y_test, y_pred_prob))
```

## Log loss[cross-entropy]
used as metric to measure the performance of classification model. It compares the probability distribution of prediction and true class labels. Lower the log loss better is the classfication model
