```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import log_loss
from sklearn.pipeline import Pipeline
```

```python
bank = pd.read_csv("/home/darkstar/Documents/pg-dbda/module7_statistics/Daywise Study Material/cases/Bankruptcy/Bankruptcy.csv")
bank
```

```python
X = bank.drop(['NO', 'D', 'YR'], axis=1)
print(X.shape)
y = bank['D']
print(y.shape)
```

```python
svm = SVC(kernel='linear', probability=True, random_state=23)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
results = cross_val_score(svm, X, y, cv=kfold,scoring='neg_log_loss')
print(results.mean())
```

```python
scaler = MinMaxScaler()
pipe = Pipeline([('SCL',scaler),('SVM', svm)])
params ={'SVM__C':np.linspace(0.001, 5, 20)}
gcv = GridSearchCV(pipe, param_grid=params, verbose=3, cv=kfold, scoring='neg_log_loss')
gcv.fit(X, y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
params = {'C':np.linspace(0.001, 5, 20), 'gamma':np.linspace(0.001, 5, 20)}
svm = SVC(kernel='rbf', probability=True, random_state=23)
gcv = GridSearchCV(svm, param_grid=params, verbose=3, cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
params = {'C':np.linspace(0.001, 5, 20), 'degree':[2,3,4]}
svm = SVC(kernel='poly', probability=True, random_state=23)
gcv = GridSearchCV(svm, param_grid=params, verbose=3, cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```
