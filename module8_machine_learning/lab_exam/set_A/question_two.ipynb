```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, ConfusionMatrixDisplay,  confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
from matplotlib import pyplot as plt
from sklearn.pipeline import Pipeline
import os
```

```python
house = pd.read_csv("california.csv")
house
```

```python
X = house.drop(['MedHouseVal'], axis = 1)
y = house['MedHouseVal']
```

## XGBRegressor

```python
clf = XGBRegressor(random_state=23)
params = {'learning_rate':[0.001, 0.002],
          'max_depth':[2,3,4,5,6,None],
          'n_estimators':[50,100,150]}
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
cv = GridSearchCV(clf, param_grid=params,cv=kfold, verbose=3, scoring='neg_mean_squared_error')

cv.fit(X,y)
df_cv = pd.DataFrame(cv.cv_results_)

print(cv.best_params_)
print(cv.best_score_)
```

conclusion: the mean squared erorr for xgb regressor is 2.28

## KNN 

```python
scaler = StandardScaler()
knn = KNeighborsRegressor()
params = {'KNN__n_neighbors': np.arange(1,31)}
pipe = Pipeline([('SCL',scaler),('KNN',knn)])
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
gcv = GridSearchCV(pipe, param_grid=params,verbose=3,
                   scoring='neg_mean_squared_error',
                   cv=kfold)
gcv.fit(X, y)
print(gcv.best_params_)
print(gcv.best_score_)
```

conclusion: the mean squared error for knn regressor is 0.39

## Random Forest

```python
rf = RandomForestRegressor(random_state=23)
kfold = KFold(n_splits=5,random_state=23,shuffle=True)
rf.get_params()

params = {'max_features': np.arange(2,6)}
gcv = GridSearchCV(rf, param_grid=params , verbose=3,cv=kfold, scoring="neg_mean_squared_error")
gcv.fit(X,y)
gcv.best_score_
gcv.best_params_
print(gcv.best_score_)
print(gcv.best_params_)
```

conclusion the mean squared error for the random forest is 0.23 which is best as compared to knn and xgb regressor.
indicating that it has the smallest discrepancy between the model's predictions and the true observations
