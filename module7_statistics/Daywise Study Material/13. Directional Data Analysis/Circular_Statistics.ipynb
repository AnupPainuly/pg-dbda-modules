```python
import numpy as np
from scipy.stats import circmean, circstd
import matplotlib.pyplot as plt
```

```python
angles = np.array([341,330,301,299,9,7,359,334,353,15,27,28,25,23,350,30,26,22,8,356])
```

```python
angles_rad = np.deg2rad(angles)
angles_rad
```

```python
cir_mean = circmean(angles_rad)
cir_mean
```

Converting from radians to degrees

```python
np.rad2deg(cir_mean)
```

```python
cir_sd = circstd(angles_rad)
cir_sd
```

Converting from radians to degrees

```python
np.rad2deg(cir_sd)
```

```python
plt.plot(np.cos(np.linspace(0, 2*np.pi, 500)),
         np.sin(np.linspace(0, 2*np.pi, 500)),
         c='lightblue')
plt.scatter(np.cos(angles_rad), np.sin(angles_rad), c='black')
plt.scatter(np.cos(cir_mean), np.sin(cir_mean), c='red', label='Circular Mean')
plt.legend()
plt.axis('equal')
plt.show()
```
